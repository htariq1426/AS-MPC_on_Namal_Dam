```python
f1 = flows1[-30:]
f2 = flows2[-30:]
f3 = flows3[-30:]
f4 = flows4[-30:]
f5 = flows5[-30:]
f6 = flows6[-30:]
f7 = flows7[-30:]
f8 = flows8[-30:]
f9 = flows9[-30:]
f10 = flows10[-30:]
```

```python
start_dt = date(1992, 9, 1)
end_dt = date(1992, 9, 30)

# difference between current and previous date
delta = timedelta(days=1)

# store the dates between two dates in a list
dr = []

while start_dt <= end_dt:
    # add current date to list by converting  it to iso format
    dr.append(start_dt)
    # increment start date by timedelta
    start_dt += delta
```

```python
ESF = pd.DataFrame({'Date':dr,'Flo_1':f1,'Flo_2':f2,'Flo_3':f3,'Flo_4':f4,'Flo_5':f5,'Flo_6':f6,'Flo_7':f7,'Flo_8':f8,'Flo_9':f9,'Flo_10':f10})
```

```python
ESF
```

```python
med = []
for i in range(30):
    dum = []
    for j in range(10):
        dum.append(ESF.iloc[i,j+1])
    med.append(np.median(np.array(dum)))
np.size(med)
```

```python
mean = []
for i in range(30):
    sum = 0.0
    for j in range(10):
        sum = sum + ESF.iloc[i,j+1]
    mean.append(sum/10)
np.size(mean)
```

```python
std = []
var = []
for i in range(30):
    sum = 0.0
    for j in range(10):
        sum = sum + (ESF.iloc[i,j+1]-mean[i])**2
    sum = sum/10
    var.append(sum)
    sum = math.sqrt(sum)
    std.append(sum)
```

```python
upper = []
upper2 = []
upper3 = []
lower = []
lower2 = []
lower3 = []
for i in range(30):
    upper.append(mean[i]+std[i])
    lower.append(max(0,mean[i]-std[i]))
    upper2.append(mean[i]+std[i]+std[i])
    lower2.append(max(0,mean[i]-std[i]-std[i]))
    upper3.append(mean[i]+std[i]+std[i]+std[i])
    lower3.append(max(0,mean[i]-std[i]-std[i]-std[i]))
```

```python
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(15)
plt.plot(ESF['Date'],ESF['Flo_1'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_2'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_3'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_4'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_5'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_6'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_7'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_8'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_9'],'--',lw=1.5)
plt.plot(ESF['Date'],ESF['Flo_10'],'--',lw=1.5)
plt.plot(ESF['Date'],mean,lw=3, label= "Ensemble Mean")
# plt.plot(dr,med,lw=3)

plt.fill_between(ESF['Date'],upper, lower, alpha=0.75, label="mean+-1*std")
plt.fill_between(ESF['Date'],upper2, lower2, alpha=0.5, label="mean+-2*std")
plt.fill_between(ESF['Date'],upper3, lower3, alpha=0.25, label="mean+-3*std")

plt.xlabel("Days of Month (Sep-1992)", fontsize=20)
plt.ylabel("Streamflows (m^3/s)", fontsize=20)
plt.title("10 Member Ensemble Streamflow Re-Forecasts for Namal Watershed (Sept-1992)", fontsize=25)

plt.xticks(rotation = 45, fontsize=15)
plt.yticks(fontsize=15)
# plt.savefig('Ensemble Memeber.jpg', dpi=500);
plt.legend(fontsize=20)
plt.show()
```

```python
pr = []
for j in range(30):
    sum = 0
    for i in range(10):
        sum = sum + ((scipy.stats.norm(mean[j],std[j]).pdf(ESF.iloc[j,i+1])))
    for i in range(10):
        pr.append((scipy.stats.norm(mean[j],std[j]).pdf(ESF.iloc[j,i+1]))/sum)
pr = np.array(pr)
```
